```python
from pathlib import Path
import pandas as pd


def find_titanic_csv() -> Path:
    for p in (Path('titanic.csv'), Path('../../data/titanic.csv'), Path('../data/titanic.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('titanic.csv не найден — положите файл рядом с ноутбуком')


TITANIC_PATH = find_titanic_csv()
df = pd.read_csv(TITANIC_PATH)

import matplotlib.pyplot as plt
from pathlib import Path as _P
_P('figures').mkdir(exist_ok=True)
```

## Урок

```python
n_rows, n_cols = df.shape
source_note = 'seaborn load_dataset(titanic) -> data/titanic.csv'
outcome_col = 'survived'
headline_rates = {
    'overall': float(df['survived'].mean()),
    'female': float(df.loc[df['sex']=='female','survived'].mean()),
    'male': float(df.loc[df['sex']=='male','survived'].mean()),
    'pclass1': float(df.loc[df['pclass']==1,'survived'].mean()),
    'pclass3': float(df.loc[df['pclass']==3,'survived'].mean()),
}
desc = df[['age','fare']].describe()
plt.figure(); df.groupby('sex')['survived'].mean().plot(kind='bar'); plt.tight_layout()
fig1 = _P('figures/report_surv_sex.png'); plt.savefig(fig1); plt.close()
plt.figure(); df.boxplot(column='age', by='pclass'); plt.suptitle(''); plt.tight_layout()
fig2 = _P('figures/report_age_pclass.png'); plt.savefig(fig2); plt.close()
MISSING_SECTION = 'age ~177 NA, deck большинство NA, embarked 2; deck не группируем; age описываем отдельно'
BIAS_SECTION = 'Подвыборки по deck/классу смещают доли; разница групп — ассоциация, не доказанная причина'
DRAFT_CONCLUSION = (
    f"Женщины выживали чаще мужчин ({headline_rates['female']:.0%} vs {headline_rates['male']:.0%}); "
    f"1 класс выше 3 ({headline_rates['pclass1']:.0%} vs {headline_rates['pclass3']:.0%}). "
    'Вывод описательный по данной таблице; модель не строилась; есть пропуски и риск bias.'
)
USED_MODEL = False
artifact_ok = {k: True for k in ('numbers','figures','missing','bias','conclusion','no_causality_overclaim')}
final_rates = dict(headline_rates)
figure_files = ['report_surv_sex.png', 'report_age_pclass.png']
REFLECTION = (
    'До модуля таблица казалась готовой к fit. Теперь сначала доли, пропуски, bias и формулировки без причинности. '
    'Выбросы fare и пустой deck показывают, что «удобный срез» врёт. '
    'Отчёт обществу — про ограничения так же, как про находки.'
)
NEXT_MODULE = 'Без EDA kNN получит мусорные признаки и смещённую оценку качества'
READY = True
print(final_rates, READY)
```
