```python
from pathlib import Path
import pandas as pd


def find_titanic_csv() -> Path:
    for p in (Path('titanic.csv'), Path('../../data/titanic.csv'), Path('../data/titanic.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('titanic.csv не найден — положите файл рядом с ноутбуком')


TITANIC_PATH = find_titanic_csv()
df = pd.read_csv(TITANIC_PATH)
```

## Урок

```python
n_f1 = int(((df['sex'] == 'female') & (df['pclass'] == 1)).sum())
rate_f1 = float(df.loc[(df['sex'] == 'female') & (df['pclass'] == 1), 'survived'].mean())
overall = float(df['survived'].mean())
REBUTTAL = f'Выжили лишь ~{overall:.1%}, не почти все'
n_child = int((df['who'] == 'child').sum())
rate_alone = float(df.loc[df['alone'] == True, 'survived'].mean())
rate_not_alone = float(df.loc[df['alone'] == False, 'survived'].mean())
by_port = df.dropna(subset=['embarked']).groupby('embarked')['survived'].agg(['count', 'mean'])
CAUSALITY_NOTE = 'Порт связан с pclass/тарифом: Cherbourg чаще 1 класс — не «магия порта».'
rate_adult_male = float(df.loc[df['adult_male'] == True, 'survived'].mean())
print(n_f1, rate_f1, REBUTTAL, n_child, by_port, rate_adult_male)
```

## ДЗ

```python
n_m3 = int(((df['sex'] == 'male') & (df['pclass'] == 3)).sum())
rate_m3 = float(df.loc[(df['sex'] == 'male') & (df['pclass'] == 3), 'survived'].mean())
top_town = str(df['embark_town'].value_counts().index[0])
rate_child = float(df.loc[df['who'] == 'child', 'survived'].mean())
rate_am = float(df.loc[df['adult_male'] == True, 'survived'].mean())
VERDICT = f'дети {rate_child:.3f} vs adult_male {rate_am:.3f} — утверждение ложно'
ct = pd.crosstab(df['sex'], df['pclass'])
n_f2 = int(ct.loc['female', 2])
print(n_m3, rate_m3, top_town, VERDICT, n_f2)
```
