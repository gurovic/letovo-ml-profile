```python
from pathlib import Path
import pandas as pd


def find_titanic_csv() -> Path:
    for p in (Path('titanic.csv'), Path('../../data/titanic.csv'), Path('../data/titanic.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('titanic.csv не найден — положите файл рядом с ноутбуком')


TITANIC_PATH = find_titanic_csv()
df = pd.read_csv(TITANIC_PATH)

import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path as _P
_P('figures').mkdir(exist_ok=True)
```

## Урок

```python
true_rate = float(df['survived'].mean())
rate_first = float(df.loc[df['pclass'] == 1, 'survived'].mean())
rate_male = float(df.loc[df['sex'] == 'male', 'survived'].mean())
BIAS_MALE = 'Мужчины выживали реже; опрос только мужчин занизит долю выживших'
rate_s50 = float(df.sample(50, random_state=0)['survived'].mean())
rate_s50_b = float(df.sample(50, random_state=1)['survived'].mean())
rate_known_deck = float(df.dropna(subset=['deck'])['survived'].mean())
LIMITATIONS = (
    'Таблица — пассажиры из учебного набора, не полный список всех на борту. '
    'Команда и часть документов отсутствуют. Срезы по deck/1 классу перепредставляют состоятельных. '
    'Нельзя переносить доли на всё население 1912 года. Вывод — об этой таблице и её ограничениях.'
)
plt.figure(); df['age'].dropna().hist(bins=20); plt.tight_layout()
plt.savefig('figures/age_hist.png'); plt.close()
AGE_SHAPE = 'возраст скошен, не идеальный колокол'
rng = np.random.default_rng(0)
fare = df['fare'].to_numpy()
means = [float(rng.choice(fare, size=40, replace=True).mean()) for _ in range(200)]
plt.figure(); plt.hist(means, bins=20); plt.tight_layout()
clt_path = _P('figures/clt_fare_means.png'); plt.savefig(clt_path); plt.close()
CLT_LIMIT = 'ЦПТ про распределение средних при больших n, не про форму сырого fare'
TOP_NA = list(df.isna().sum().sort_values(ascending=False).head(3).index)
n_drop = int(df['age'].notna().sum())
age_filled_median = df['age'].fillna(df['age'].median())
ZERO_BAD = 'Возраст 0 путается с младенцами и ломает mean/модель'
n_all_drop = int(len(df.dropna()))
DROP_ALL_RISK = 'Остаются в основном строки с известным deck — смещение к 1 классу'
CLEAN_LOG = (
    'deck: не импутируем (слишком много NA). age: для описательных таблиц — median fill или анализ с NA отдельно. '
    'duplicates: drop_duplicates перед агрегатами. embarked: 2 NA — можно dropna точечно.'
)
STRATA_IDEA = 'Страты по pclass сохраняют структуру классов и меньше тянут bias'
print(true_rate, TOP_NA, n_all_drop)
```
