# Sampling bias; ЦПТ; пропуски

Три темы одной пары: смещённая подвыборка, колокол средних, стратегия чистки.

```python
from pathlib import Path
import pandas as pd


def find_titanic_csv() -> Path:
    for p in (Path('titanic.csv'), Path('../../data/titanic.csv'), Path('../data/titanic.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('titanic.csv не найден — положите файл рядом с ноутбуком')


TITANIC_PATH = find_titanic_csv()
df = pd.read_csv(TITANIC_PATH)

import matplotlib.pyplot as plt
import numpy as np
```

## 1. Истинная доля выживших

`true_rate = df['survived'].mean()` — эталон «полной» таблицы урока.

```python
true_rate = float(df['survived'].mean())
assert 0.3 < true_rate < 0.5
print(round(true_rate, 3))
```

## 2. Смещённые подвыборки

`rate_first` (pclass==1) и `rate_male`. Запишите `BIAS_MALE` — почему опрос только мужчин врёт.

```python
rate_first = None
rate_male = None
BIAS_MALE = ''
assert rate_first is not None and float(rate_first) > true_rate
assert rate_male is not None and float(rate_male) < true_rate
assert len(BIAS_MALE) > 30
print(round(float(rate_first), 3), round(float(rate_male), 3), BIAS_MALE)
```

## 3. Случайная выборка n=50

Два `sample(50)` с `random_state` 0 и 1 → `rate_s50`, `rate_s50_b`.
