```python
from pathlib import Path
import pandas as pd


def find_titanic_csv() -> Path:
    for p in (Path('titanic.csv'), Path('../../data/titanic.csv'), Path('../data/titanic.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('titanic.csv не найден — положите файл рядом с ноутбуком')


TITANIC_PATH = find_titanic_csv()
df = pd.read_csv(TITANIC_PATH)
```

## Урок. 1. Осмотр

```python
n_rows, n_cols = df.shape
print(n_rows, n_cols)
print(df.head())
print(list(df.columns))
```

## Урок. 2. survival_rate

```python
survival_rate = float(df['survived'].mean())
print('доля выживших:', round(survival_rate, 3))
```

## Урок. 6. График

```python
import matplotlib.pyplot as plt
from pathlib import Path as _P
figures_dir = _P('figures')
figures_dir.mkdir(exist_ok=True)
ax = df['pclass'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('pclass'); plt.ylabel('count'); plt.tight_layout()
saved_path = figures_dir / 'pclass_counts.png'
plt.savefig(saved_path); plt.close()
print('saved:', saved_path)
```
