# Выборка, репрезентативность, sampling bias

Подвыборка может врать о «всех пассажирах». Учимся ловить bias.

```python
from pathlib import Path
import pandas as pd


def find_titanic_csv() -> Path:
    for p in (Path('titanic.csv'), Path('../../data/titanic.csv'), Path('../data/titanic.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('titanic.csv не найден — положите файл рядом с ноутбуком')


TITANIC_PATH = find_titanic_csv()
df = pd.read_csv(TITANIC_PATH)
```

## 1. Истинная доля выживших

`true_rate = df['survived'].mean()` — эталон «полной» таблицы урока.

```python
true_rate = float(df['survived'].mean())
assert 0.3 < true_rate < 0.5
print(round(true_rate, 3))
```

## 2. Смещённая подвыборка: только 1 класс

Доля выживших среди `pclass == 1` → `rate_first`. Насколько выше `true_rate`?

```python
rate_first = None
assert rate_first is not None and isinstance(rate_first, float)
assert rate_first > true_rate
print(round(rate_first, 3), 'vs', round(true_rate, 3))
```

## 3. Смещённая подвыборка: только мужчины

`rate_male` vs `true_rate`. Запишите `BIAS_MALE` — почему общество ошибётся, если опросит только мужчин.

```python
rate_male = None
BIAS_MALE = ''
assert rate_male is not None and float(rate_male) < true_rate
assert len(BIAS_MALE) > 30
print(round(float(rate_male), 3), BIAS_MALE)
```

## 4. Случайная выборка n=50

`df.sample(50, random_state=0)['survived'].mean()` → `rate_s50`. Повторите с `random_state=1` → `rate_s50_b`. Разброс?

```python
rate_s50 = None
rate_s50_b = None
assert rate_s50 is not None and rate_s50_b is not None
print(rate_s50, rate_s50_b, abs(float(rate_s50) - float(rate_s50_b)))
```

## 5. Репрезентативность

Сравните долю женщин во всей таблице и в `pclass==1`. → `share_f_all`, `share_f_first`.

**Идея:** bias по классу тянет за собой bias по полу.

```python
share_f_all = None
share_f_first = None
assert share_f_all is not None and share_f_first is not None
assert float(share_f_first) > float(share_f_all)
print(round(float(share_f_all), 3), round(float(share_f_first), 3))
```
