# Среднее, медиана, std — вручную и в pandas

Описываем `age` и `fare`: центр и разброс. Модель не нужна.

```python
from pathlib import Path
import pandas as pd


def find_titanic_csv() -> Path:
    for p in (Path('titanic.csv'), Path('../../data/titanic.csv'), Path('../data/titanic.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('titanic.csv не найден — положите файл рядом с ноутбуком')


TITANIC_PATH = find_titanic_csv()
df = pd.read_csv(TITANIC_PATH)

ages = df['age'].dropna()
fares = df['fare']
```

## 1. Среднее age вручную

`sum(ages) / len(ages)` → `mean_age_manual`. Сверьте с `ages.mean()`.

```python
mean_age_manual = None
mean_age_pd = float(ages.mean())
assert mean_age_manual is not None
assert abs(float(mean_age_manual) - mean_age_pd) < 1e-9
print(mean_age_manual, mean_age_pd)
```

## 2. Медиана age

Упорядочите значения и возьмите середину (при чётном n — среднее двух центральных). Или `ages.median()`. → `median_age`.

**Why:** медиана устойчивее к выбросам, чем mean.

```python
median_age = None
assert median_age is not None
assert abs(float(median_age) - float(ages.median())) < 1e-9
print(median_age)
```

## 3. Std age

`ages.std()` (выборочное, ddof=1 в pandas по умолчанию). → `std_age`.

**Интуиция:** типичный разброс вокруг среднего в тех же единицах (годы).

```python
std_age = None
assert std_age is not None and isinstance(float(std_age), float)
assert 10 < float(std_age) < 20
print(round(float(std_age), 2))
```

## 4. mean vs median для fare

Сравните `fares.mean()` и `fares.median()`. Почему mean больше? Запишите в `FARE_NOTE`.

```python
fare_mean = None
fare_median = None
FARE_NOTE = ''
assert fare_mean is not None and fare_median is not None
assert float(fare_mean) > float(fare_median)
assert len(FARE_NOTE) > 20
print(round(float(fare_mean), 2), round(float(fare_median), 2), FARE_NOTE)
```

## 6. Эксперимент: выброс

Скопируйте `fares` в список/Series, добавьте искусственный билет 5000, пересчитайте mean и median. Что сдвинулось сильнее? → `OUTLIER_EFFECT`.

```python
mean_with_outlier = None
median_with_outlier = None
OUTLIER_EFFECT = ''
assert mean_with_outlier is not None and median_with_outlier is not None
assert float(mean_with_outlier) > float(fares.mean())
assert abs(float(median_with_outlier) - float(fares.median())) < 1.0
assert len(OUTLIER_EFFECT) > 15
print(mean_with_outlier, median_with_outlier, OUTLIER_EFFECT)
```
