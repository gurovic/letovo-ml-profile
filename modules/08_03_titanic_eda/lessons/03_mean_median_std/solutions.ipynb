```python
from pathlib import Path
import pandas as pd


def find_titanic_csv() -> Path:
    for p in (Path('titanic.csv'), Path('../../data/titanic.csv'), Path('../data/titanic.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('titanic.csv не найден — положите файл рядом с ноутбуком')


TITANIC_PATH = find_titanic_csv()
df = pd.read_csv(TITANIC_PATH)

ages = df['age'].dropna()
fares = df['fare']
```

## Урок

```python
mean_age_manual = float(sum(ages) / len(ages))
median_age = float(ages.median())
std_age = float(ages.std())
fare_mean = float(fares.mean())
fare_median = float(fares.median())
FARE_NOTE = 'mean тянут редкие очень дорогие билеты; медиана ближе к типичному'
fare_q3 = float(fares.quantile(0.75))
fares2 = pd.concat([fares, pd.Series([5000.0])])
mean_with_outlier = float(fares2.mean())
median_with_outlier = float(fares2.median())
OUTLIER_EFFECT = 'mean вырос сильно, median почти нет'
age_by_class = df.groupby('pclass')['age'].mean()
WHEN_MEDIAN = 'Когда есть дорогие выбросы и нужен типичный билет, а не среднее арифметическое'
print(mean_age_manual, median_age, std_age, fare_q3, age_by_class)
```

## ДЗ

```python
a_mean, a_med, a_std = float(ages.mean()), float(ages.median()), float(ages.std())
fare_by_sex = df.groupby('sex')['fare'].mean()
thr = df['fare'].quantile(0.99)
mean_after = float(df.loc[df['fare'] <= thr, 'fare'].mean())
MEAN_DROP_NOTE = f'mean {df["fare"].mean():.1f} -> {mean_after:.1f} после отсечения top 1%'
toy_median = 3
print(a_mean, fare_by_sex, MEAN_DROP_NOTE, toy_median)
```
