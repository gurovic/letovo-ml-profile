```python
from pathlib import Path
import pandas as pd


def find_titanic_csv() -> Path:
    for p in (Path('titanic.csv'), Path('../../data/titanic.csv'), Path('../data/titanic.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('titanic.csv не найден — положите файл рядом с ноутбуком')


TITANIC_PATH = find_titanic_csv()
df = pd.read_csv(TITANIC_PATH)

import matplotlib.pyplot as plt
from pathlib import Path as _P
_P('figures').mkdir(exist_ok=True)
```

## Урок

```python
ages = df['age'].dropna()
q1, q2, q3 = float(ages.quantile(0.25)), float(ages.quantile(0.5)), float(ages.quantile(0.75))
iqr_age = q3 - q1
upper_fence = q3 + 1.5 * iqr_age
plt.figure(); df.boxplot(column='age'); plt.tight_layout()
age_box_path = _P('figures/age_box.png'); plt.savefig(age_box_path); plt.close()
plt.figure(); df.boxplot(column='fare', by='pclass'); plt.suptitle(''); plt.tight_layout()
fare_box_path = _P('figures/fare_by_pclass.png'); plt.savefig(fare_box_path); plt.close()
fq1, fq3 = float(df['fare'].quantile(0.25)), float(df['fare'].quantile(0.75))
n_fare_outliers = int((df['fare'] > fq3 + 1.5 * (fq3 - fq1)).sum())
age_med_by_who = df.groupby('who')['age'].median()
WRONG_CLAIM = 'Boxplot fare не содержит survived — нельзя вывести выживаемость из одного разброса цены'
plt.figure(); df.boxplot(column='age', by='survived'); plt.suptitle(''); plt.tight_layout()
age_surv_path = _P('figures/age_by_survived.png'); plt.savefig(age_surv_path); plt.close()
AGE_SURV_NOTE = 'медианы близки; возраст сам по себе слабый разделитель'
print(q1, q2, q3, n_fare_outliers, age_med_by_who)
```
