## 7. Fibonacci — антипример

Наивная рекурсия vs итерация для n=25. Вывод: pipeline — итерация; рекурсия — для **структур данных**.

```python
def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n - 1) + fib_recursive(n - 2)


def fib_iterative(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a


# import time
# сравните время для n=25
```
