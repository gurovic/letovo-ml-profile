## 1. map — нормализация списка оценок

```python
raw = EXAM_SCORES
mx = max(raw)
scaled = list(map(lambda x: x / mx, raw))
print(scaled)
```

## 2. filter — отбор «аномалий» (score < 50 или > 90)

```python
anomalies = list(filter(lambda s: s < 50 or s > 90, EXAM_SCORES))
print(anomalies)
```

## 3. sorted — ранжирование моделей по метрике

```python
models = [
    {'name': 'knn', 'f1': 0.81},
    {'name': 'logreg', 'f1': 0.77},
    {'name': 'baseline', 'f1': 0.65},
]

ranked = sorted(models, key=lambda m: m['f1'], reverse=True)
print([m['name'] for m in ranked])
```

## 4. Практика: min_max_scale через map

```python
def min_max_scale_list(values):
    lo, hi = min(values), max(values)
    return list(map(lambda x: (x - lo) / (hi - lo), values))


print(min_max_scale_list([r['area_sqm'] for r in FEATURE_ROWS]))
```
