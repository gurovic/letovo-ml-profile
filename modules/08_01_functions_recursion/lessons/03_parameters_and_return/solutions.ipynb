```python
EXAM_SCORES = [40, 55, 62, 75, 88, 91, 48, 100, 33, 67]
AREAS_SQM = [28, 32, 45, 55, 60]
ROW_AREAS = [28, 45, 60, 32]
ROW_ROOMS = [1, 2, 3, 1]
ROW_FLOORS = [3, 7, 12, 2]
```

## Урок. 1. `describe_numbers`

```python
def describe_numbers(values):
    if not values:
        return (0.0, 0.0, 0.0, 0)
    return (sum(values) / len(values), min(values), max(values), len(values))


assert describe_numbers([10, 20, 30])[0] == 20
```

## Урок. 2. `min_max_scale`

```python
def min_max_scale(values):
    if not values:
        return []
    lo, hi = min(values), max(values)
    if lo == hi:
        return [0.0] * len(values)
    return [(x - lo) / (hi - lo) for x in values]


scaled = min_max_scale(AREAS_SQM)
assert abs(min(scaled) - 0) < 1e-9 and abs(max(scaled) - 1) < 1e-9
```

## Урок. 3. `clip_outlier`

```python
def clip_outlier(x, low, high):
    return max(low, min(high, x))
```

## Урок. 4. mutable default

```python
def collect_outliers_ok(x, bucket=None):
    if bucket is None:
        bucket = []
    if x > 100:
        bucket.append(x)
    return bucket


assert collect_outliers_ok(105) == [105]
assert collect_outliers_ok(110) == [110]
```

## Урок. 5. `grade_stats`

```python
def grade_stats(scores, pass_threshold=60):
    passed = sum(1 for s in scores if s >= pass_threshold)
    failed = len(scores) - passed
    mean = sum(scores) / len(scores) if scores else 0.0
    return (mean, passed, failed)


mean, passed, failed = grade_stats(EXAM_SCORES)
assert passed + failed == len(EXAM_SCORES)
```

## ДЗ. 2. Clip всей выборки

```python
def clip_scores(scores, low, high):
    return [clip_outlier(s, low, high) for s in scores]


before = describe_numbers(EXAM_SCORES)
clipped = clip_scores(EXAM_SCORES, 50, 95)
after = describe_numbers(clipped)
print('mean до:', before[0], 'после:', after[0])
```

## ДЗ. 3. `row_numeric_stats`

```python
def row_numeric_stats(area, rooms, floor):
    nums = [area, rooms, floor]
    return (sum(nums) / len(nums), max(nums))


assert abs(row_numeric_stats(ROW_AREAS[0], ROW_ROOMS[0], ROW_FLOORS[0])[0] - 10.666666) < 1e-5
```

## ДЗ. 4. Подбор порога

```python
best = None
for threshold in range(40, 96):
    _, passed, failed = grade_stats(EXAM_SCORES, threshold)
    rate = passed / len(EXAM_SCORES)
    diff = abs(rate - 0.5)
    if best is None or diff < best[0]:
        best = (diff, threshold, rate)
print(f'лучший порог = {best[1]}, доля сдавших = {best[2]:.2f}')
```
