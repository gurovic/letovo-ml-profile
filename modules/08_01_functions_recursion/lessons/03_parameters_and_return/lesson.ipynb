## 3. Clip выброса

Ограничьте одно число границами `low` и `high`. Сравните mean до и после clip для одного «дикого» балла.

```python
def clip_outlier(x, low, high):
    return max(low, min(high, x))


wild = EXAM_SCORES[7]  # 100
print('до clip:', describe_numbers(EXAM_SCORES))
clipped_one = clip_outlier(wild, 0, 95)
print('clip(100, 0, 95) =', clipped_one)
```

## 4. Опасный default

Запустите `collect_outliers_bad` дважды. Почему список растёт? Исправьте через `bucket=None`.

```python
def collect_outliers_bad(x, bucket=[]):
    if x > 100:
        bucket.append(x)
    return bucket


print(collect_outliers_bad(105))
print(collect_outliers_bad(110))


def collect_outliers_ok(x, bucket=None):
    if bucket is None:
        bucket = []
    if x > 100:
        bucket.append(x)
    return bucket


assert collect_outliers_ok(105) == [105]
assert collect_outliers_ok(110) == [110]
```
