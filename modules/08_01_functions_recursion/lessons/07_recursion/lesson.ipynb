## 3. Fibonacci — антипример

Сравните время `fib_recursive(25)` и `fib_iterative(25)`. Рекурсия уместна для **структур** (списки, деревья), а не для «просто посчитать число».

```python
import time


def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n - 1) + fib_recursive(n - 2)


def fib_iterative(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a


N = 25
t0 = time.perf_counter()
fib_recursive(N)
t_rec = time.perf_counter() - t0
t0 = time.perf_counter()
fib_iterative(N)
t_iter = time.perf_counter() - t0
print(f'recursive: {t_rec:.3f}s, iterative: {t_iter:.3f}s')
assert fib_recursive(10) == fib_iterative(10)
```
