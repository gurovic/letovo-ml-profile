## 3. `clip_outlier` — ограничение выброса

```python
def clip_outlier(x, low, high):
    return max(low, min(high, x))


print(clip_outlier(150, 0, 100))
print(clip_outlier(EXAM_SCORES[0], 0, 100))
```

## 4. Опасный default: `bucket=[]`

```python
def collect_outliers_bad(x, bucket=[]):
    if x > 100:
        bucket.append(x)
    return bucket


print(collect_outliers_bad(105))
print(collect_outliers_bad(110))  # почему список растёт?
```

```python
def collect_outliers_ok(x, bucket=None):
    if bucket is None:
        bucket = []
    if x > 100:
        bucket.append(x)
    return bucket
```
