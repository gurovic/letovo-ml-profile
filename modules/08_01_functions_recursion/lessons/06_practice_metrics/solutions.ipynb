# Решения: практика метрик

**Для преподавателя.**

```python
PREDICTIONS = [1, 0, 1, 1, 0, 1, 1, 0, 1, 0]
LABELS =      [1, 0, 1, 0, 0, 1, 1, 0, 1, 1]
```

## Урок. confusion_counts

```python
def confusion_counts(preds, labels):
    tp = fp = fn = tn = 0
    for p, y in zip(preds, labels):
        if p == 1 and y == 1:
            tp += 1
        elif p == 1 and y == 0:
            fp += 1
        elif p == 0 and y == 1:
            fn += 1
        elif p == 0 and y == 0:
            tn += 1
    return tp, fp, fn, tn


assert confusion_counts(PREDICTIONS[:4], LABELS[:4]) == (2, 1, 0, 1)
```

## ДЗ. counter

```python
def increment(counter):
    return counter + 1


def reset_and_count(items):
    total = 0
    for x in items:
        total += x
    return total


c = 0
c = increment(c)
assert c == 1
assert reset_and_count([1, 2, 3]) == 6
```
