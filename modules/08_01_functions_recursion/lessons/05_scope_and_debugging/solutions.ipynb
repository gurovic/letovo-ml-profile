# Решения: scope и отладка

**Для преподавателя.** Все задачи из `lesson.ipynb` и `homework.ipynb`.

```python
PREDICTIONS = [1, 0, 1, 1, 0, 1, 1, 0, 1, 0]
LABELS =      [1, 0, 1, 0, 0, 1, 1, 0, 1, 1]
SCORES = [72, 55, 88, 44, 61, 90, 77, 48, 83, 58]
```

## Урок. 1–2. accuracy / my_accuracy

```python
def accuracy(preds, labels):
    if len(preds) != len(labels):
        return None
    correct = sum(p == y for p, y in zip(preds, labels))
    return correct / len(labels)


REF_ACCURACY = accuracy(PREDICTIONS, LABELS)


def my_accuracy(preds, labels):
    if len(preds) != len(labels):
        return None
    if not preds:
        return 0.0
    correct = 0
    for p, y in zip(preds, labels):
        if p == y:
            correct += 1
    return correct / len(labels)


assert abs(my_accuracy(PREDICTIONS, LABELS) - REF_ACCURACY) < 1e-9
```

## Урок. 3. accuracy_fixed

```python
def accuracy_fixed(preds, labels):
    correct = 0
    for i in range(len(preds)):
        if preds[i] == labels[i]:
            correct += 1
    return correct / len(preds) if preds else 0.0
```

## Урок. 4. count_correct

```python
def count_correct(preds, labels):
    total = 0
    for p, y in zip(preds, labels):
        if p == y:
            total += 1
    return total


assert count_correct(PREDICTIONS, LABELS) == 7
```

## Урок. 5. predict_pass

```python
def predict_pass(score, threshold=60):
    return 1 if score >= threshold else 0


assert predict_pass(72) == 1
```

## ДЗ. 2. batch_predict_pass

```python
def batch_predict_pass(scores, threshold=60):
    return [predict_pass(s, threshold) for s in scores]


preds = batch_predict_pass(SCORES, 60)
assert len(preds) == len(SCORES)
```

## ДЗ. 4. shadowing

```python
def accuracy_shadow_bug(preds, labels):
    labels = 0  # затёрли аргумент!
    correct = sum(p == labels for p in preds)
    return correct / len(preds)


def accuracy_shadow_ok(preds, labels):
    correct = sum(p == y for p, y in zip(preds, labels))
    return correct / len(preds)


assert abs(accuracy_shadow_ok(PREDICTIONS, LABELS) - REF_ACCURACY) < 1e-9
```
