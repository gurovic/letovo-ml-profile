## 1. Эталон accuracy

Accuracy = доля верных предсказаний (`pred == label`). Прочитайте эталон и вычислите значение на `PREDICTIONS` / `LABELS`.

```python
def accuracy(preds, labels):
    if len(preds) != len(labels):
        return None
    correct = sum(p == y for p, y in zip(preds, labels))
    return correct / len(labels)


REF_ACCURACY = accuracy(PREDICTIONS, LABELS)
print('эталон accuracy =', REF_ACCURACY)
```
