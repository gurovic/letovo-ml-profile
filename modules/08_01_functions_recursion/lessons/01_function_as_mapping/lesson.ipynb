## 1. Проблема: дублирование предсказаний

Три квартиры — один и тот же код прогноза цены.

```python
def predict_price_v0(area_sqm):
    return 1.5 + 0.09 * area_sqm


for apt in APARTMENTS:
    p = 1.5 + 0.09 * apt['area_sqm']  # дублирование формулы
    print(apt['area_sqm'], '→', p, 'млн')
```

## 2. Модель как функция

Линейная модель (как в будущем sklearn, но **коэффициенты заданы**):

`price = 1.5 + 0.09 × area`

```python
def predict_price(area_sqm):
    """Предсказать цену квартиры (млн руб.) по площади."""
    return PRICE_INTERCEPT + PRICE_COEF_AREA * area_sqm


for apt in APARTMENTS:
    pred = predict_price(apt['area_sqm'])
    print(f"{apt['area_sqm']} м² → {pred:.2f} млн (факт: {apt['price_mln']})")
```

## 3. return vs print в ML-коде

`return` — предсказание можно **использовать дальше** (метрика, pipeline).

`print` — только вывод в консоль.

```python
def predict_print(area):
    print(PRICE_INTERCEPT + PRICE_COEF_AREA * area)


def predict_return(area):
    return PRICE_INTERCEPT + PRICE_COEF_AREA * area


a = predict_print(45)
print('a =', a)  # None

b = predict_return(45)
print('b =', b)  # число

errors = [abs(predict_return(x['area_sqm']) - x['price_mln']) for x in APARTMENTS]
print('MAE (упрощённо):', sum(errors) / len(errors))
```

**Рефлексия:** зачем в ML-инференсе нужен именно `return`?
