# ДЗ: осмотр таблицы

```python
from pathlib import Path
import pandas as pd


def find_trips_csv() -> Path:
    for p in (Path('trips.csv'), Path('../../data/trips.csv'), Path('../data/trips.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('trips.csv не найден')


TRIPS_PATH = find_trips_csv()
df = pd.read_csv(TRIPS_PATH)
```

## 1. Срез столбцов

Выведите только `distance_km`, `duration_min`, `hour` — первые 10 строк.

```python
subset = df[['distance_km', 'duration_min', 'hour']]
print(subset.head(10))
```

## 2. Одна поездка

Найдите строку с `trip_id == 'T0001'` и выведите длительность.

```python
row = df[df['trip_id'] == 'T0001']
assert len(row) == 1
print(row['duration_min'].iloc[0])
```

## 3. Сколько поездок

Посчитайте число строк таблицы двумя способами: `len(df)` и `df.shape[0]`.

```python
assert len(df) == df.shape[0]
print(len(df))
```
