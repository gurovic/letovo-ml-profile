# ДЗ: осмотр таблицы

```python
from pathlib import Path
import pandas as pd


def find_trips_csv() -> Path:
    for p in (Path('trips.csv'), Path('../../data/trips.csv'), Path('../data/trips.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('trips.csv не найден')


TRIPS_PATH = find_trips_csv()
df = pd.read_csv(TRIPS_PATH)
```

## 1. Срез столбцов

Первые 10 строк только `distance_km`, `duration_min`, `hour`.

```python
subset = None  # DataFrame 10×3
assert subset is not None
assert list(subset.columns) == ['distance_km', 'duration_min', 'hour']
assert len(subset) == 10
print(subset)
```

## 2. Одна поездка

Длительность поездки с `trip_id == 'T0001'`.

```python
duration_t0001 = None  # число
assert duration_t0001 == df.loc[df['trip_id'] == 'T0001', 'duration_min'].iloc[0]
print(duration_t0001)
```

## 3. Два способа длины

Проверьте, что `len(df)` и `df.shape[0]` совпадают.

```python
a = len(df)
b = df.shape[0]
assert a == b
print(a)
```

## 4. Вызов

Средняя `duration_min` только для `vehicle_type == 'comfort'` (округлите до 1 знака). Сравните со средним по всей таблице — одной фразой.

```python
mean_comfort = None  # float
mean_all = round(df['duration_min'].mean(), 1)
COMPARE_NOTE = ''  # одна фраза
assert mean_comfort is not None
assert abs(mean_comfort - round(df.loc[df['vehicle_type'] == 'comfort', 'duration_min'].mean(), 1)) < 1e-9
assert len(COMPARE_NOTE) > 10
print(mean_comfort, mean_all, COMPARE_NOTE)
```
