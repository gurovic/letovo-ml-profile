```python
from pathlib import Path
import pandas as pd


def find_trips_csv() -> Path:
    for p in (Path('trips.csv'), Path('../../data/trips.csv'), Path('../data/trips.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('trips.csv не найден')


TRIPS_PATH = find_trips_csv()
df = pd.read_csv(TRIPS_PATH)
```

## Урок. 1. Первая таблица

```python
n_rows, n_cols = df.shape
print(n_rows, n_cols)
print(df.head())
print(list(df.columns))
assert n_rows == len(df)
assert n_cols == df.shape[1]
```

## Урок. 2. Признак / цель / id

```python
TARGET = 'duration_min'
FEATURES = ['distance_km', 'hour', 'zone', 'vehicle_type']
ID_COL = 'trip_id'
assert TARGET == 'duration_min'
assert ID_COL not in FEATURES
```

## Урок. 3–4. object и зоны

```python
object_cols = df.select_dtypes(include='object').columns.tolist()
n_zones = df['zone'].nunique()
print(object_cols, n_zones)
```

## ДЗ. 1–4

```python
subset = df[['distance_km', 'duration_min', 'hour']].head(10)
duration_t0001 = df.loc[df['trip_id'] == 'T0001', 'duration_min'].iloc[0]
mean_comfort = round(df.loc[df['vehicle_type'] == 'comfort', 'duration_min'].mean(), 1)
mean_all = round(df['duration_min'].mean(), 1)
COMPARE_NOTE = f'comfort: {mean_comfort}, все: {mean_all} — сравните порядок величин'
print(subset)
print(duration_t0001, mean_comfort, COMPARE_NOTE)
```
