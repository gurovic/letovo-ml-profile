```python
from pathlib import Path
import pandas as pd


def find_listings_csv() -> Path:
    for p in (Path('listings.csv'), Path('../../data/listings.csv'), Path('../data/listings.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('listings.csv не найден')


LISTINGS_PATH = find_listings_csv()
df = pd.read_csv(LISTINGS_PATH)
```

## Урок. 1. Зачем таблица

```python
n_rows, n_cols = df.shape
print(n_rows, n_cols)
print(df.head())
print(list(df.columns))
```

## Урок. 2. dtypes

```python
accommodates_dtype = str(df['accommodates'].dtype)
neighbourhood_dtype = str(df['neighbourhood'].dtype)
print('accommodates:', accommodates_dtype, '| neighbourhood:', neighbourhood_dtype)
```

## Урок. 3. Одна ячейка

```python
first_accommodates = float(df.iloc[0]['accommodates'])
print('первый listing, accommodates:', first_accommodates)
```

## Урок. 4. loc vs iloc

```python
by_name = df.loc[:2, ['accommodates', 'price']]
by_pos = df.iloc[:3][['accommodates', 'price']]
assert (by_name.values == by_pos.values).all()
print(by_name)
print(by_pos)
```

## Урок. 5. Признак / цель / id

```python
TARGET = 'price'
FEATURES = ['accommodates', 'number_of_reviews', 'neighbourhood', 'room_type']
ID_COL = 'listing_id'
assert TARGET not in FEATURES and ID_COL not in FEATURES
```

## Урок. 6. value_counts

```python
neighbourhood_counts = df['neighbourhood'].value_counts()
n_neighbourhoods = int(df['neighbourhood'].nunique())
print('уникальных neighbourhood:', n_neighbourhoods)
print(neighbourhood_counts)
```

## ДЗ. 4. Средняя entire

```python
mean_entire = round(df.loc[df['room_type'] == 'entire', 'price'].mean(), 1)
mean_all = round(df['price'].mean(), 1)
COMPARE_NOTE = f'entire: {mean_entire}, все: {mean_all} — сравните порядок величин'
print(mean_entire, mean_all, COMPARE_NOTE)
```
