```python
from pathlib import Path
import pandas as pd


def find_trips_csv() -> Path:
    for p in (Path('trips.csv'), Path('../../data/trips.csv'), Path('../data/trips.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('trips.csv не найден')


TRIPS_PATH = find_trips_csv()
df = pd.read_csv(TRIPS_PATH)
```

## Урок. 1. Зачем таблица

```python
n_rows, n_cols = df.shape
print(n_rows, n_cols)
print(df.head())
print(list(df.columns))
```

## Урок. 2. dtypes

```python
distance_dtype = str(df['distance_km'].dtype)
zone_dtype = str(df['zone'].dtype)
print('distance:', distance_dtype, '| zone:', zone_dtype)
```

## Урок. 3. Одна ячейка

```python
first_distance = float(df.iloc[0]['distance_km'])
print('первая поездка, км:', first_distance)
```

## Урок. 4. loc vs iloc

```python
by_name = df.loc[:2, ['distance_km', 'duration_min']]
by_pos = df.iloc[:3][['distance_km', 'duration_min']]
assert (by_name.values == by_pos.values).all()
print(by_name)
print(by_pos)
```

## Урок. 5. Признак / цель / id

```python
TARGET = 'duration_min'
FEATURES = ['distance_km', 'hour', 'zone', 'vehicle_type']
ID_COL = 'trip_id'
assert TARGET not in FEATURES and ID_COL not in FEATURES
```

## Урок. 6. value_counts

```python
zone_counts = df['zone'].value_counts()
n_zones = int(df['zone'].nunique())
print('уникальных zone:', n_zones)
print(zone_counts)
```

## ДЗ. 4. Средняя comfort

```python
mean_comfort = round(df.loc[df['vehicle_type'] == 'comfort', 'duration_min'].mean(), 1)
mean_all = round(df['duration_min'].mean(), 1)
COMPARE_NOTE = f'comfort: {mean_comfort}, все: {mean_all} — сравните порядок величин'
print(mean_comfort, mean_all, COMPARE_NOTE)
```
