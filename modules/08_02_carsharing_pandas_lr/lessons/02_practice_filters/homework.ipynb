# ДЗ: фильтры

```python
from pathlib import Path
import pandas as pd


def find_trips_csv() -> Path:
    for p in (Path('trips.csv'), Path('../../data/trips.csv'), Path('../data/trips.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('trips.csv не найден')


TRIPS_PATH = find_trips_csv()
df = pd.read_csv(TRIPS_PATH)
```

### A. Закрепление

## 1. Комфорт

Только `vehicle_type == 'comfort'` — число строк.

```python
n_comfort = None
assert n_comfort == (df['vehicle_type'] == 'comfort').sum()
print(n_comfort)
```

## 2. Утро

`hour` от 7 до 9 включительно — число строк.

```python
n_morning = None
assert n_morning == ((df['hour'] >= 7) & (df['hour'] <= 9)).sum()
print(n_morning)
```

## 3. Топ-5

5 самых длинных по `duration_min` (`trip_id`, `duration_min`).

```python
top5 = None
assert top5 is not None
assert len(top5) == 5
assert top5['duration_min'].is_monotonic_decreasing
print(top5[['trip_id', 'duration_min']])
```
