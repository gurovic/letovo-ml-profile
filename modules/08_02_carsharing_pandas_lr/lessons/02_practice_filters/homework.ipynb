# ДЗ: фильтры

```python
from pathlib import Path
import pandas as pd


def find_trips_csv() -> Path:
    for p in (Path('trips.csv'), Path('../../data/trips.csv'), Path('../data/trips.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('trips.csv не найден')


TRIPS_PATH = find_trips_csv()
df = pd.read_csv(TRIPS_PATH)
```

## 1. Комфорт

Только `vehicle_type == 'comfort'`.

```python
comfort = df[df['vehicle_type'] == 'comfort']
print(len(comfort))
```

## 2. Утро

Поездки с `hour` от 7 до 9 включительно.

```python
morning = df[(df['hour'] >= 7) & (df['hour'] <= 9)]
print(len(morning))
```

## 3. Сортировка

5 самых длинных по `duration_min`.

```python
top5 = df.sort_values('duration_min', ascending=False).head(5)
print(top5[['trip_id', 'duration_min']])
```
