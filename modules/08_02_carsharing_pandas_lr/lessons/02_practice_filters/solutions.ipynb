```python
from pathlib import Path
import pandas as pd


def find_trips_csv() -> Path:
    for p in (Path('trips.csv'), Path('../../data/trips.csv'), Path('../data/trips.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('trips.csv не найден')


TRIPS_PATH = find_trips_csv()
df = pd.read_csv(TRIPS_PATH)
```

## Урок. 1. Фильтр center

```python
center = df[df['zone'] == 'center']
print('center:', len(center), 'из', len(df))
```

## Урок. 2. Длинные поездки

```python
long_trips = df[df['distance_km'] >= 15]
print(long_trips[['trip_id', 'distance_km', 'duration_min']].head())
```

## Урок. 3. Составная маска

```python
center_long = df[(df['zone'] == 'center') & (df['distance_km'] >= 10)]
print(len(center_long))
```

## Урок. 4. Отрицание ~

```python
not_center = df[~(df['zone'] == 'center')]
assert len(not_center) + len(center) == len(df)
print('not center:', len(not_center))
```

## Урок. 5. Топ-3

```python
top3 = df.sort_values('duration_min', ascending=False).head(3)
print(top3[['trip_id', 'duration_min']])
```

## Урок. 6. Доля center

```python
share_center = round(len(center) / len(df), 3)
print('доля center:', share_center)
```

## Урок. 7. loc vs iloc после фильтра

```python
first_center_id = str(center.iloc[0]['trip_id'])
print(first_center_id)
```

## Урок. 8. Тип hour

```python
hours = df['hour'].astype(int)
print(int(hours.min()), int(hours.max()))
```

## ДЗ. 1. Комфорт

```python
n_comfort = int((df['vehicle_type'] == 'comfort').sum())
print(n_comfort)
```

## ДЗ. 2. Утро

```python
n_morning = int(((df['hour'] >= 7) & (df['hour'] <= 9)).sum())
print(n_morning)
```

## ДЗ. 3. Топ-5

```python
top5 = df.sort_values('duration_min', ascending=False).head(5)
print(top5[['trip_id', 'duration_min']])
```

## ДЗ. 4. Доля center

```python
share_center = round((df['zone'] == 'center').mean(), 3)
SHARE_NOTE = f'center ≈ {share_center:.0%} поездок — ориентир доли центра'
print(share_center, SHARE_NOTE)
```

## ДЗ. 5. Не comfort

```python
n_not_comfort = int((df['vehicle_type'] != 'comfort').sum())
print(n_not_comfort)
```
