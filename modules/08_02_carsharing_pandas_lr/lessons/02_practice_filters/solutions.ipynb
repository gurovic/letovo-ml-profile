```python
from pathlib import Path
import pandas as pd


def find_listings_csv() -> Path:
    for p in (Path('listings.csv'), Path('../../data/listings.csv'), Path('../data/listings.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('listings.csv не найден')


LISTINGS_PATH = find_listings_csv()
df = pd.read_csv(LISTINGS_PATH)
```

## Урок. 1. Фильтр center

```python
center = df[df['neighbourhood'] == 'center']
print('center:', len(center), 'из', len(df))
```

## Урок. 2. Вместительные объявления

```python
large_listings = df[df['accommodates'] >= 6]
print(large_listings[['listing_id', 'accommodates', 'price']].head())
```

## Урок. 3. Составная маска

```python
center_long = df[(df['neighbourhood'] == 'center') & (df['accommodates'] >= 4)]
print(len(center_long))
```

## Урок. 4. Отрицание ~

```python
not_center = df[~(df['neighbourhood'] == 'center')]
assert len(not_center) + len(center) == len(df)
print('not center:', len(not_center))
```

## Урок. 5. Топ-3

```python
top3 = df.sort_values('price', ascending=False).head(3)
print(top3[['listing_id', 'price']])
```

## Урок. 6. Доля center

```python
share_center = round(len(center) / len(df), 3)
print('доля center:', share_center)
```

## Урок. 7. loc vs iloc после фильтра

```python
first_center_id = str(center.iloc[0]['listing_id'])
print(first_center_id)
```

## Урок. 8. Тип number_of_reviews

```python
reviews = df['number_of_reviews'].astype(int)
print(int(reviews.min()), int(reviews.max()))
```

## ДЗ. 1. Entire

```python
n_entire = int((df['room_type'] == 'entire').sum())
print(n_entire)
```

## ДЗ. 2. Отзывы 10–50

```python
n_mid_reviews = int(((df['number_of_reviews'] >= 7) & (df['number_of_reviews'] <= 9)).sum())
print(n_mid_reviews)
```

## ДЗ. 3. Топ-5

```python
top5 = df.sort_values('price', ascending=False).head(5)
print(top5[['listing_id', 'price']])
```

## ДЗ. 4. Доля center

```python
share_center = round((df['neighbourhood'] == 'center').mean(), 3)
SHARE_NOTE = f'center ≈ {share_center:.0%} объявлений — ориентир доли центра'
print(share_center, SHARE_NOTE)
```

## ДЗ. 5. Не entire

```python
n_not_entire = int((df['room_type'] != 'entire').sum())
print(n_not_entire)
```
