# EDA: describe и scatter

```python
from pathlib import Path
import pandas as pd


def find_trips_csv() -> Path:
    for p in (Path('trips.csv'), Path('../../data/trips.csv'), Path('../data/trips.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('trips.csv не найден')


TRIPS_PATH = find_trips_csv()
df = pd.read_csv(TRIPS_PATH)
import matplotlib.pyplot as plt
```

## 1. describe

Описательная статистика для `distance_km` и `duration_min`.

```python
print(df[['distance_km', 'duration_min']].describe())
```

## 2. info

`df.info()` — сколько пропусков?

```python
df.info()
```

## 3. Scatter

Постройте scatter: ось X — `distance_km`, ось Y — `duration_min`. Подпишите оси и заголовок.

```python
plt.figure(figsize=(6, 4))
plt.scatter(df['distance_km'], df['duration_min'], alpha=0.5)
plt.xlabel('distance_km')
plt.ylabel('duration_min')
plt.title('Длительность vs расстояние')
plt.show()
```
