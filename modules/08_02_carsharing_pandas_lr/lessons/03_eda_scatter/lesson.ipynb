Перед `fit` аналитик **смотрит** данные: типичные значения, пропуски, есть ли связь признака с целью. Это EDA — exploratory data analysis.

```python
from pathlib import Path
import pandas as pd


def find_trips_csv() -> Path:
    for p in (Path('trips.csv'), Path('../../data/trips.csv'), Path('../data/trips.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('trips.csv не найден')


TRIPS_PATH = find_trips_csv()
df = pd.read_csv(TRIPS_PATH)
import matplotlib.pyplot as plt
```

## 4. Scatter: подписи и ловушки

Scatter `distance_km` → `duration_min`.

**Pitfalls:**
- без `xlabel`/`ylabel`/`title` график «немой»;
- при тысячах точек облако заливает — `alpha=0.4…0.6`;
- оси перепутаны → ложный вывод о связи.

Допишите подписи, затем `SCATTER_DONE = True`.

```python
plt.figure(figsize=(6, 4))
plt.scatter(df['distance_km'], df['duration_min'], alpha=0.5)
# plt.xlabel(...); plt.ylabel(...); plt.title(...)
plt.show()
SCATTER_DONE = False
assert SCATTER_DONE
```

## 6. hour — другой признак

Медиана `hour` и scatter `hour` → `duration_min` (с подписями). Сравните с distance: облако плотнее или размытее?

```python
median_hour = None
assert median_hour is not None
assert 0 <= float(median_hour) <= 23
plt.figure(figsize=(6, 4))
plt.scatter(df['hour'], df['duration_min'], alpha=0.4)
plt.xlabel('hour')
plt.ylabel('duration_min')
plt.title('Длительность vs час')
plt.show()
print('median hour:', median_hour)
```

## 7. Корреляция — число, не доказательство

`corr_dist` — корреляция Пирсона между `distance_km` и `duration_min`.

**Важно:** corr близко к 1 говорит о **линейной** согласованности на этих данных, но **не доказывает** причинность («далеко → долго из‑за X»). Запишите число и фразу-ограничение в `CORR_NOTE`.

```python
corr_dist = None
CORR_NOTE = ''
assert corr_dist is not None
assert -1 <= float(corr_dist) <= 1
assert len(CORR_NOTE) > 20
print('corr(distance, duration):', round(float(corr_dist), 3))
print(CORR_NOTE)
```
