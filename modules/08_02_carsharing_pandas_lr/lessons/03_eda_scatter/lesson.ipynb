## 4. Scatter: подписи и ловушки

Scatter `accommodates` → `price`.

**Pitfalls:**
- без `xlabel`/`ylabel`/`title` график «немой»;
- при тысячах точек облако заливает — `alpha=0.4…0.6`;
- оси перепутаны → ложный вывод о связи.

Допишите подписи, затем `SCATTER_DONE = True`.

```python
plt.figure(figsize=(6, 4))
plt.scatter(df['accommodates'], df['price'], alpha=0.5)
# plt.xlabel(...); plt.ylabel(...); plt.title(...)
plt.show()
SCATTER_DONE = False
assert SCATTER_DONE
```

## 6. number_of_reviews — другой признак

Медиана `number_of_reviews` и scatter `number_of_reviews` → `price` (с подписями). Сравните с accommodates: облако плотнее или размытее?

```python
median_reviews = None
assert median_reviews is not None
assert 0 <= float(median_reviews) <= 23
plt.figure(figsize=(6, 4))
plt.scatter(df['number_of_reviews'], df['price'], alpha=0.4)
plt.xlabel('number_of_reviews')
plt.ylabel('price')
plt.title('Цена vs отзывы')
plt.show()
print('median reviews:', median_reviews)
```

## 7. Корреляция — число, не доказательство

`corr_acc` — корреляция Пирсона между `accommodates` и `price`.

**Важно:** corr близко к 1 говорит о **линейной** согласованности на этих данных, но **не доказывает** причинность («далеко → долго из‑за X»). Запишите число и фразу-ограничение в `CORR_NOTE`.

```python
corr_acc = None
CORR_NOTE = ''
assert corr_acc is not None
assert -1 <= float(corr_acc) <= 1
assert len(CORR_NOTE) > 20
print('corr(accommodates, price):', round(float(corr_acc), 3))
print(CORR_NOTE)
```
