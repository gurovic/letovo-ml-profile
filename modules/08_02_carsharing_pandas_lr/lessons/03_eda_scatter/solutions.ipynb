```python
from pathlib import Path
import pandas as pd


def find_trips_csv() -> Path:
    for p in (Path('trips.csv'), Path('../../data/trips.csv'), Path('../data/trips.csv')):
        if p.exists():
            return p.resolve()
    raise FileNotFoundError('trips.csv не найден')


TRIPS_PATH = find_trips_csv()
df = pd.read_csv(TRIPS_PATH)
import matplotlib.pyplot as plt
```

## Урок. 1. describe

```python
stats = df[['distance_km', 'duration_min']].describe()
print(stats)
mean_duration = float(df['duration_min'].mean())
print('mean duration:', mean_duration)
```

## Урок. 2. mean vs median

```python
median_duration = float(df['duration_min'].median())
MEAN_MEDIAN_NOTE = (
    'Сравните mean и median: если mean выше — длинный хвост поездок тянет среднее.'
)
print(mean_duration, median_duration, MEAN_MEDIAN_NOTE)
```

## Урок. 3. info

```python
n_non_null_duration = int(df['duration_min'].notna().sum())
print('non-null duration:', n_non_null_duration, '/', len(df))
```

## Урок. 4. Scatter distance

```python
plt.figure(figsize=(6, 4))
plt.scatter(df['distance_km'], df['duration_min'], alpha=0.5)
plt.xlabel('distance_km')
plt.ylabel('duration_min')
plt.title('Длительность vs расстояние')
plt.show()
SCATTER_DONE = True
```

## Урок. 5. Вывод EDA

```python
EDA_CONCLUSION = (
    'С ростом distance_km длительность в среднем растёт, но разброс заметный.'
)
print(EDA_CONCLUSION)
```

## Урок. 6. hour scatter

```python
median_hour = float(df['hour'].median())
plt.figure(figsize=(6, 4))
plt.scatter(df['hour'], df['duration_min'], alpha=0.4)
plt.xlabel('hour')
plt.ylabel('duration_min')
plt.title('Длительность vs час')
plt.show()
print('median hour:', median_hour)
```

## Урок. 7. Корреляция

```python
corr_dist = float(df['distance_km'].corr(df['duration_min']))
CORR_NOTE = (
    'corr — мера линейной согласованности на этих данных, не доказательство причины.'
)
print('corr(distance, duration):', round(corr_dist, 3))
print(CORR_NOTE)
```

## Урок. 8. Checkpoint признак

```python
BETTER_FOR_LINEAR = 'distance_km'
WHY_EDA = (
    'На scatter с distance видна наклонная полоса; у hour облако более размытое. '
    'Линейная модель на distance — первый кандидат.'
)
print(BETTER_FOR_LINEAR, WHY_EDA)
```
